```python
def download_prev_data(companies):
    bse = nyse.valid_days(start_date = dt.date.today()-dt.timedelta(days=10), end_date=dt.date.today()-dt.timedelta(days=1))
    bse = bse[-6:]
    dataset = pdr.get_data_yahoo(companies, 
                                  #start = str(dt.date.today()-dt.timedelta(days=5)),
                                  start    = bse[0],
                                  end      = bse[-1],
                                  interval = "5m",
                                  group_by ='ticker'
                                 )
    return dataset
def download_today_data(companies):
    dataset = pdr.get_data_yahoo(companies,
                                  start = "2021-02-12",
                                  #start = str(dt.date.today()),
                                  interval = "5m",
                                  group_by ='ticker'
                                 )
    return dataset
data  = download_prev_data(c)
```

```python
def avg_cumulative_vol(c,av_data):
    vol = av_data[c]["Volume"]
    x = np.empty((5,75), dtype=int)
    for i in range(0,5):
        x[i][:] =vol[75*i : 75*i+75]
    y = x.cumsum(axis=1)
    y = y.sum(axis=0)/5
    return y
def cumulative_vol(c,data):
    vol = data[c]["Volume"]
    vol = vol[0:75].values
    y   = vol.cumsum()
    return y
```

```python
today_data = download_today_data(c)
def check(c):
    for i in c:
        k =np.zeros(2)
        x=avg_cumulative_vol(i,data)
        y=cumulative_vol(i,today_data)
        
        y = np.nan_to_num(y)
        if y.shape[0]<75:
            k = np.zeros(75-y.shape[0])
            y = np.append(y,k)
        relative_vol = y/x
        print("\n\n")
        print("The relative volume of:",i)
        print(relative_vol)
    return relative_vol
```
